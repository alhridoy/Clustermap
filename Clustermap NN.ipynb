```python
new_df['LAT']=pd.Series(new_df['lat'], dtype="string").tolist()
new_df['LON']=pd.Series(new_df['lon'], dtype="string").tolist()
```

```python
from folium.plugins import MarkerCluster # for clustering the markers
m = folium.Map(location=[35.854390,-111.596281], default_zoom_start=12)
m.choropleth(geo_data='nn.json', # I found this NYC zipcode boundaries by googling 
             data=geo, # my dataset
             columns=['OBJECTID', 'pop'], # zip code is here for matching the geojson zipcode, sales price is the column that changes the color of zipcode areas
             key_on='feature.properties.OBJECTID', # this path contains zipcodes in str type, this zipcodes should match with our ZIP CODE column
             fill_color='BuPu', fill_opacity=0.7, line_opacity=0.3,
             legend_name='population')

# add a marker for every record in the filtered data, use a clustered view
marker_cluster = MarkerCluster().add_to(m) # create marker clusters
for i in range(geo.shape[0]):
        location=list([new_df['LON'][i],new_df['LAT'][i]])
        
   
        #tooltip = "Zipcode:{}<br> Borough: {}<br> Click for more".format(geo["OBJECTID"][i], geo['Chapter'][i])
        #print(tooltip)
        folium.Marker(location,
                
                  popup="""<i>Mean sales price: </i> <br> <b>${}</b> <br>
                  <i>mean : </i><b><br>{}</b><br>
                  <i>mean : </i><b><br>{}</b><br>"""
                  .format(round(geo['pop'][i],2),
                  round(geo['pctPop'][i],2), round(geo['LSpop'][i],2))
                  ).add_to(marker_cluster)
m
```
